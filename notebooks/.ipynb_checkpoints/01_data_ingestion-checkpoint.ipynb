```python
# Load nav history
nav = pd.read_csv("../data/raw/02_nav_history.csv")

# Load fund master
master = pd.read_csv("../data/raw/01_fund_master.csv")

# Check: all codes in nav_history should exist in fund_master
nav_codes = set(nav["amfi_code"].unique())
master_codes = set(master["amfi_code"].unique())

missing = nav_codes - master_codes

if len(missing) == 0:
    print("✅ All AMFI codes match! No issues found.")
else:
    print(f"❌ {len(missing)} codes in NAV not found in master:")
    print(missing)
```
