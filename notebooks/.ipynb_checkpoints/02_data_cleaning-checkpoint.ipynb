```python
import pandas as pd
import numpy as np
import os

# Load nav_history
nav = pd.read_csv("../data/raw/02_nav_history.csv")

print("BEFORE CLEANING:")
print(f"Shape: {nav.shape}")
print(f"Columns: {nav.columns.tolist()}")
print(f"Missing values:\n{nav.isnull().sum()}")
print(nav.head(3))
```

```python
# Step 1: Parse date column to datetime
nav["date"] = pd.to_datetime(nav["date"])

# Step 2: Sort by amfi_code + date
nav = nav.sort_values(["amfi_code", "date"]).reset_index(drop=True)

# Step 3: Remove duplicates
nav = nav.drop_duplicates(subset=["amfi_code", "date"])

# Step 4: Remove invalid NAV (NAV must be > 0)
nav = nav[nav["nav"] > 0]

# Step 5: Forward fill missing NAV for weekends/holidays
nav = nav.set_index("date")
nav = nav.groupby("amfi_code").apply(
    lambda x: x.resample("D").ffill()
).reset_index(level=0, drop=True).reset_index()

# Step 6: Remove weekends (Saturday=5, Sunday=6)
nav = nav[nav["date"].dt.dayofweek < 5]

print("AFTER CLEANING:")
print(f"Shape: {nav.shape}")
print(f"Missing values: {nav.isnull().sum().sum()}")
print(nav.head(3))

# Save cleaned file
nav.to_csv("../data/processed/clean_nav.csv", index=False)
print("✅ clean_nav.csv saved!")
```

```python
# Load transactions
tx = pd.read_csv("../data/raw/08_investor_transactions.csv")

print("BEFORE CLEANING:")
print(f"Shape: {tx.shape}")
print(f"Transaction types: {tx['transaction_type'].unique()}")
print(f"KYC status: {tx['kyc_status'].unique()}")
```

```python
# Step 1: Fix date format
tx["transaction_date"] = pd.to_datetime(tx["transaction_date"])

# Step 2: Standardise transaction_type
tx["transaction_type"] = tx["transaction_type"].str.strip().str.title()
# Makes sure values are: Sip, Lumpsum, Redemption

# Step 3: Remove invalid amounts
tx = tx[tx["amount_inr"] > 0]

# Step 4: Fix KYC status
tx["kyc_status"] = tx["kyc_status"].str.strip().str.title()

# Step 5: Remove duplicates
tx = tx.drop_duplicates()

print("AFTER CLEANING:")
print(f"Shape: {tx.shape}")
print(f"Transaction types: {tx['transaction_type'].unique()}")
print(f"Missing values: {tx.isnull().sum().sum()}")

# Save
tx.to_csv("../data/processed/clean_transactions.csv", index=False)
print("✅ clean_transactions.csv saved!")
```

```python
# Load performance
perf = pd.read_csv("../data/raw/07_scheme_performance.csv")

print("BEFORE CLEANING:")
print(f"Shape: {perf.shape}")
print(f"Columns: {perf.columns.tolist()}")
print(perf.head(3))
```

```python
# Step 1: Check expense ratio range (should be 0.1 to 2.5)
print("Expense ratio range:")
print(f"Min: {perf['expense_ratio_pct'].min()}")
print(f"Max: {perf['expense_ratio_pct'].max()}")

# Flag anomalies
anomalies = perf[
    (perf["expense_ratio_pct"] < 0.1) | 
    (perf["expense_ratio_pct"] > 2.5)
]
print(f"\nAnomalies found: {len(anomalies)}")

# Step 2: Check negative Sharpe ratios
negative_sharpe = perf[perf["sharpe_ratio"] < 0]
print(f"Funds with negative Sharpe: {len(negative_sharpe)}")

# Step 3: Make sure numeric columns are numeric
numeric_cols = ["return_1yr_pct", "return_3yr_pct", 
                "return_5yr_pct", "sharpe_ratio", "alpha", "beta"]
for col in numeric_cols:
    perf[col] = pd.to_numeric(perf[col], errors="coerce")

# Save
perf.to_csv("../data/processed/clean_performance.csv", index=False)
print("✅ clean_performance.csv saved!")
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Create database connection
engine = create_engine(
    "sqlite:///../data/db/bluestock_mf.db"
)

# Load all cleaned files
fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav = pd.read_csv("../data/processed/clean_nav.csv")
tx = pd.read_csv("../data/processed/clean_transactions.csv")
perf = pd.read_csv("../data/processed/clean_performance.csv")
aum = pd.read_csv("../data/raw/03_aum_by_fund_house.csv")

# Load into database
fund_master.to_sql("dim_fund", engine, 
                   if_exists="replace", index=False)
nav.to_sql("fact_nav", engine, 
           if_exists="replace", index=False)
tx.to_sql("fact_transactions", engine, 
          if_exists="replace", index=False)
perf.to_sql("fact_performance", engine, 
            if_exists="replace", index=False)
aum.to_sql("fact_aum", engine, 
           if_exists="replace", index=False)

print("✅ All tables loaded into bluestock_mf.db!")

# Verify row counts
for table in ["dim_fund", "fact_nav", 
              "fact_transactions", "fact_performance", "fact_aum"]:
    result = pd.read_sql(f"SELECT COUNT(*) as count FROM {table}", 
                         engine)
    print(f"{table}: {result['count'][0]} rows")
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "sqlite:///../data/db/bluestock_mf.db"
)

# Test Query 1 — Top 5 funds by AUM
print("TOP 5 FUNDS BY AUM:")
result = pd.read_sql("""
    SELECT fund_house, MAX(aum_crore) as max_aum
    FROM fact_aum
    GROUP BY fund_house
    ORDER BY max_aum DESC
    LIMIT 5
""", engine)
print(result)

# Test Query 3 — Transaction types
print("\nTRANSACTION TYPES:")
result2 = pd.read_sql("""
    SELECT transaction_type,
           COUNT(*) as count,
           SUM(amount_inr) as total_amount
    FROM fact_transactions
    GROUP BY transaction_type
""", engine)
print(result2)
```
